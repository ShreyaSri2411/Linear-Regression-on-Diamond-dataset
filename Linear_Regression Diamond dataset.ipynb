```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dia=sns.load_dataset('diamonds')
dia.head()
```

```python
dia.shape
```

```python
diaf=dia[
    dia.cut.isin(['Premium','Very Good']) &
    dia.color.isin(['D','F','J'])
].sample(n=200)
```

```python
diaf
```

```python
df=diaf.loc[:,['carat','price']]
pd.set_option('display.max_rows',None)
df
```

```python
sns.set_style('dark')
sns.scatterplot(x='carat',y='price',data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
w=lr.coef_
b=lr.intercept_
```

```python
0.90*w+b
```

```python
x_test.head()
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
sns.set_style('dark')
sns.scatterplot(x='carat',y='price',data=df)
plt.plot(x_test,y_pred,color='red')
plt.xlabel('carat')
plt.ylabel('price')
plt.show()
```

```python
lr.score(x_test,y_test)
```
